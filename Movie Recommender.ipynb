```python
import ast
```

```python
def convert_dict_to_list(obj) :

    L = []
    # ast is importent to covert to list
    for i in ast.literal_eval(obj):

        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert_dict_to_list)
```

```python
movies['keywords'] = movies['keywords'].apply(convert_dict_to_list)
```

```python
def convert_Multip_dict_to_list(obj) :

    L = []
    counter = 0
    # ast is importent to covert to list
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert_Multip_dict_to_list)
```

```python
movies['cast'] = movies['cast'].apply(convert_Multip_dict_to_list)
```

```python
def mult_dict_fetch_director(obj) :

    L = []
    # ast is importent to covert to list
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(mult_dict_fetch_director)
```

```python
import nltk

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
# Ex

ps.stem('dansing')
```

```python
def stem(text):

    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```
